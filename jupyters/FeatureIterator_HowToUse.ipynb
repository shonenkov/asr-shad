```python
import matplotlib.pyplot as plt
import sys
import pandas as pd
import os
import torchaudio

sys.path.insert(0, '../utils')
from word_iterators import word_wave_iterator, word_feature_iterator
from audio_stats import get_audio_stats
from int_to_text import num2text

from IPython.display import Audio
```

```python
resources_path = '../numbers'
```

```python
df = pd.read_csv(os.path.join(resources_path, 'train.csv'))
df['filename'] = df.path.apply(os.path.basename)
df_audio = get_audio_stats(os.path.join(resources_path, 'train'))
df = df_audio.merge(df, on='filename')
df['text'] = df.number.apply(num2text)
df['num_words'] = df['text'].apply(lambda t: len(t.split(' ')))
df.head(3)
```

# dummy slicing with n_words (from train)

```python
iterator = word_feature_iterator(resources_path, df)
```

```python
word, word_wave, mfcc, gender = next(iterator)
torchaudio.save('word_{}.wav'.format(word), word_wave, sample_rate=word_wave.shape[1])
print('word:', word)
print('gender:', gender)
print('MFCC shape:', mfcc.shape)
plt.imshow(mfcc[0, :, :].numpy(), cmap='nipy_spectral')
plt.show()
Audio('word_{}.wav'.format(word))
```
