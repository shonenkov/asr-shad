```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

import sys
sys.path.insert(0, '..')
from utils.text_stats import *
```

```python
df = pd.read_csv('../numbers/train.csv')
df.head(3)
```

### words

```python
counter = get_word_counter(df.number)
# 15 most common words
most_common = counter.most_common(15)
fig = plt.figure(figsize=(20,5))
plt.bar([x[0] for x in most_common], [x[1] for x in most_common])
plt.show()
```

### word count

```python
counter = get_wordcnt_counter(df.number)
plt.bar(counter.keys(), counter.values())
plt.show()
```

### chars

```python
counter = get_char_counter(df.number)
plt.bar(counter.keys(), counter.values())
plt.show()
```

### digits

```python
counter = get_digit_counter(df.number)
plt.bar(counter.keys(), counter.values())
plt.show()
```

### digit count

```python
counter = get_digitcnt_counter(df.number)
plt.bar(counter.keys(), counter.values())
plt.show()
```

### word cloud

```python
text = ' '.join(get_word_pool(df.number))
wordcloud = WordCloud().generate(text)
fig = plt.figure(figsize=(15,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
